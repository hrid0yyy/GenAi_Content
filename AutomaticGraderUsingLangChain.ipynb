```python
from langchain_community.llms import HuggingFacePipeline
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import torch

model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

hf_pipeline = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=100,
    device=0 if torch.cuda.is_available() else -1
)

llm = HuggingFacePipeline(pipeline=hf_pipeline)

template_text = """You are an automatic grader. Decide whether the student's answer is correct or incorrect based on the question and correct answer. Simple misspellings are okay.

Question: {question}
Correct Answer: {correct_answer}
Student Answer: {student_answer}

Is the student answer correct? Answer "Correct" or "Incorrect" only.
"""

prompt_template = PromptTemplate(
    input_variables=["question", "correct_answer", "student_answer"],
    template=template_text
)

grader_chain = LLMChain(llm=llm, prompt=prompt_template)

question = "What is the primary source of energy for Earth's climate system?"
correct_answer = "The Sun"
student_answer = "The Sun is the main energy source for Earth's climate, powering weather and ocean movements."

result = grader_chain.run(
    question=question,
    correct_answer=correct_answer,
    student_answer=student_answer
)

print(result.strip())  # Should output: "Correct"
```
